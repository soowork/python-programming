```python
# 1. clean the data by dropping columns I won't use
# use inplace so that I change my actual data frame
df.drop(['bookID','isbn','isbn13'],axis = 1,inplace=True) 
```

```python
# 2A. I only want to know about English language books
# which language codes are those?
langs = pd.DataFrame(df['language_code'].value_counts().reset_index())
langs.columns = ['language','count']
print(langs[(langs['language'].str[:2]) == 'en'])                  
```

```python
# 2B. filter the data frame - this time I will create a new df instead of 
#      working in place
#    Note that 'enm' is middle english - I won't include those!
# NOTE - using a set here
en_codes = {'en-CA','en-GB','en-US','eng'}
books_en = df[df.language_code.isin(en_codes)]
books_en.head(10)
```

```python
# we have filtered the data from to 13719 to 12648
books_en.shape
```

```python
# 3. data prep -- make sure I have a rating for everything 
# returns nothing - so looks good
books_en[books_en['average_rating'].isnull()]
```

```python
# 4. data prep -- I want to exclude books that are scifi or fantasy or vampires 'cause I don't enjoy those genres
# using a set for this
exclude = {'J.K. Rowling','J.K. Rowling-Mary GrandPré','J.R.R. Tolkien','George R.R. Martin','George R.R. Martin-Lauren Cannon','J.R.R. Tolkien-Peter S. Beagle','Diana Gabaldon','Elie Wiesel-Marion Wiesel-François Mauriac','J.R. Ward','Orson Scott Card','Isaac Asimov','Jim Butcher','Robin Hobb','Tamora Pierce','Laurell K. Hamilton','Dr. Seuss'}
# new df books_good will be the books excluding those authors
books_good = books_en[~books_en.authors.isin(exclude)]
```

```python
# filtered further from 12648 to 12451
books_good.shape
```

```python
# it is common to get books with very high reviews, but not many reviewers
# the mean count of text reviews was 533.439901 (see [16])
# the books I want to read have rating > 4.1 and have more than 50K reviews

readem = books_good[(books_good['average_rating'] > 4.1) & (books_good['ratings_count'] > 50000)].sort_values('average_rating',ascending=False)
```

```python
# will utilize a class to do a little data adjustment
# since I only care about ratings over 4 - might as well make them from 0-1
class CleanUp:
    
    def adjust_rating(rating):
        return (rating - 4)
```

```python
# add a new column to the data frame for this adjusted rating 
# note that when I pass the average_rating to the function I am using keyword argument, not positional 
readem['adjusted_rating'] = CleanUp.adjust_rating(rating=readem['average_rating'])
```

```python
# looks good
readem.head(10)
```

```python
# what should Sue read next 
rated = readem.sort_values('ratings_count',ascending=False).head(30).set_index('title')
ply.subplots(figsize=(10,9))
ax = rated['adjusted_rating'].sort_values().plot.barh(width=0.9,color=sns.color_palette('muted'))
ax.set_xlabel("Adjusted Rating (Value over 4)")
ax.set_ylabel("Books")
ax.set_title("What Should Sue Read Next")
ply.show()
```
