```python
# read in a dump of book & review data from the Goodreads (https://www.kaggle.com/jealousleopard/goodreadsbooks)
df = pd.read_csv('books.csv')
```

```python
df.columns
```

```python
# 1. clean the data by dropping columns I won't use
# use inplace so that I change my actual data frame
df.drop(['bookID','isbn','isbn13'],axis = 1,inplace=True) 
```

```python
# 2A. I only want to know about English language books
# which language codes are those?
df['language_code'].value_counts().head()
```

```python
# 2B. filter the data frame - this time I will create a new df instead of 
#      working in place
#   en-CA        9
#   en-GB      341
#   en-US     1700
#   eng      10598
#   enm          3 -- middle english - no thanks!

en_codes = ['en-CA','en-GB','en-US','eng']
books_en = df[df.language_code.isin(en_codes)]
```

```python
# we have filtered the data from to 13719 to 12648
books_en.shape
```

```python
books_en.head(10)
```

```python
# 3. data prep -- make sure I have a rating for everything 
# returns nothing - so looks good
books_en[books_en['average_rating'].isnull()]
```

```python
# I want to exclude books that are scifi or fantasy cause I don't enjoy those genres
exclude = ['J.K. Rowling','J.K. Rowling-Mary GrandPré','J.R.R. Tolkien','George R.R. Martin','George R.R. Martin-Lauren Cannon','J.R.R. Tolkien-Peter S. Beagle','Diana Gabaldon','Elie Wiesel-Marion Wiesel-François Mauriac','J.R. Ward','Orson Scott Card','Isaac Asimov']
```

```python
# books_en[books_en.authors in exclude]
books_good = books_en[~books_en.authors.isin(exclude)]
```

```python
books_good.shape
```

```python
# it is common to get books with very high reviews, but not many reviewers
# the mean count of text reviews was 533.439901 (see [16])
# I am curious about what books are > 4.1 and have more reviews than 533.439901

# works
# books_good[books_good.average_rating > 4.3].sort_values('ratings_count',ascending=False)

books_good[(books_good['average_rating'] > 4.1) & (books_good['ratings_count'] > 100000)].sort_values('ratings_count',ascending=False)
```

```python
readem = books_good[(books_good['average_rating'] > 4.3) & (books_good['ratings_count'] > 50000)].sort_values('average_rating',ascending=False)
```

```python
# next ideas - filter out Tolkien, Rowling, Martin, Gabaldon, J.R. Ward, etc
# use text_reviews_count because those are the folks that REALLY care!
```

```python
most_rated = df.sort_values('ratings_count',ascending=False).head(10).set_index('title')
plt.subplots(figsize=(12,8))
ax = most_rated['ratings_count'].sort_values().plot.barh(width=0.9,color=sns.color_palette('CMRmap',12))
ax.set_xlabel("Total ratings count ", fontsize=15)
ax.set_ylabel("Books", fontsize=15)
ax.set_title("Top 10 most rated books",fontsize=20,color='black')
totals = []
for i in ax.patches:
    totals.append(i.get_width())
total = sum(totals)
for i in ax.patches:
    ax.text(i.get_width()+3, i.get_y()+.2,str(round(i.get_width())), fontsize=15,color='black')
plt.show()
```
